# Problem Statement

* ***  Zomato dataset gives a list of restaurants ,their dining types and user ratings ,location of restaurants in the city and the services offered by each restaurant listed in the dataset and also it gives a list of URLs of restaurants for online orders .This project is to  perfom Exploratory Data Analysis on given Zomato dataset so that the dataset is available for further analysis or any predictions of a feature to train a machine learning model ***

# Features

* *** 1)url:identifies restaurant online ***
* *** 2)address:Physical location of restaurant ***
* *** 3)name:Name of the restaurant ***
* *** 4)online_order: checks whether online order available or not ***
* *** 5)book_table:checks whether customer can reserve  a table ***
* *** 6)rate:user ratings scaled for 5 ***
* *** 7)votes:tells how many users were rated this restaurant ***
* *** 8)phone:landline/mob no. of restaurant ***
* *** 9)location :place where  restaurant is located ***
* *** 10)rest_type:identifies what type of restaurant ie., whether casual dining or bar ***
* *** 11)dish_liked:gives a list of famous dishes identified from user ratings ***
* *** 12)cuisines:type of cuisine restaurant is offering ***
* *** 13)approx_cost(for two people):tells approximate cost for 2 people ***
* *** 14)reviews_list=customer ratings given in specific catagory ***
* *** 15)'menu_item':food and beverages menu available with the restaurant ***
* *** 16)listed_in(type):also identifies type of restaurant ***
* *** 17)listed_in(city):also identifies location within the city ***

# Data cleaning

```python
# Dropping unnecessary columns
df=df.drop(columns=["url","address","reviews_list","menu_item","dish_liked","cuisines"])
df
```

# drop duplicates

dup=df[df.duplicated()]
df.drop_duplicates(inplace=True)

```python
df
```

```python
# fixing name column
# df["name"]=df["name"].str.replace('','e',regex=False)
```

```python
# standardizing mobile number
df['phone']=df['phone'].str.replace(' ','',regex=False)
df['phone']=df['phone'].str.extract(r'(\+91\d{10}|0\d{10})')
df['phone']=df['phone'].str.extract(r'(?:\+91|0)?(\d{10})')
```

```python
# cleaning rate column
df['rate'] = df['rate'].str.replace('NEW', '', regex = False)
df['rate'] = df['rate'].str.extract('([0-9]\.[0-9])')
df['rate'] = df['rate'].fillna(0).astype(float)
```

```python
# checking whether to standardize book_table column
print(df.online_order.unique())
print(df.book_table.unique())
```

```python
# Fixing online_order and book_table for further eda
df["online_order"]=df["online_order"].str.replace('Yes','1',regex=False)
df["online_order"]=df["online_order"].str.replace('No','0',regex=False)
df["book_table"]=df["book_table"].str.replace('Yes','1',regex=False)
df["book_table"]=df["book_table"].str.replace('No','0',regex=False)
```

```python
# Standardizing cost column
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(' ','',regex=False)
df['approx_cost(for two people)']=df['approx_cost(for two people)'].str.replace(',','',regex=False)
```

# Handling missing values


```python
# filling missing phone numbers with others phone numbers with same name
df['phone']=df.groupby('name')['phone'].transform(lambda x:x.ffill().bfill())
```

```python
df['location']=df.groupby('name')['location'].transform(lambda x:x.ffill().bfill())
df['rest_type']=df.groupby('name')['rest_type'].transform(lambda x:x.ffill().bfill())
df['approx_cost(for two people)']=df.groupby('name')['approx_cost(for two people)'].transform(lambda x:x.ffill().bfill())
```

```python
df['rest_type'].fillna('Not available',inplace=True)
df['approx_cost(for two people)'].fillna('0',inplace=True)
```

```python
# dropping row
df.dropna(subset={'phone'},inplace=True)
```

```python
# converting datatype
df['rate']=df['rate'].astype('float')
df['approx_cost(for two people)']=df['approx_cost(for two people)'].astype('int64')
df['online_order']=df['online_order'].astype('int')
df['book_table']=df['book_table'].astype('int')
```
